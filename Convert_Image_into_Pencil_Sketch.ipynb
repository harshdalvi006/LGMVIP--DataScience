# **Importing Required Libraries**

```python
import cv2
from PIL import Image
from IPython.display import display
```

```python
original_img = cv2.imread('/content/photo.jpg')
```

```python
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(original_img))
```

# **Converting the Original Image into Grayscale**

```python
gray_image = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

# **Inverting the Grayscale image**

```python
inv_gray_image = 255 - gray_image
display(Image.fromarray(inv_gray_image))
```

```python

blur_image = cv2.GaussianBlur(inv_gray_image, (35, 35), 0)
```

```python
display(Image.fromarray(blur_image))
```

# **Inverting the Blurred Image**

```python
inverted_blur_image = 255 - blur_image
```

# **Creating the Pencil Sketch image**


```python
pencil_sketch = cv2.divide(gray_image, inverted_blur_image, scale=256.0)
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(pencil_sketch)
```

We need to read the image and then convert it to a grayscale image. This will turn an image into a black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image.

```python
display(Image.fromarray(original_img))
display(Image.fromarray(pencil_sketch))
```
