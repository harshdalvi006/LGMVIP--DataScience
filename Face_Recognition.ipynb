**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')
from mtcnn import MTCNN  # helps in detecting faces from image
from tensorflow.keras.models import load_model
```

```python
model = load_model('/content/model.h5')
print('Model Loaded 100%')
```

**FOR ANGRY**

```python
img = '/content/Faces/angry/PrivateTest_1054527.jpg'
try:
    image = plt.imread(img)
except SyntaxError:
    image = plt.imread(img, format='jpg')
plt.imshow(image)
```

```python
test_image = np.resize(image, (90,3,1))
```

```python
label_dict = {0:'Angry',1:'Disgust',2:'Fear',3:'Happy',4:'Neutral',5:'Sad',6:'Surprise'}
test_img = np.expand_dims(test_image, axis = 0)
result = model.predict(test_img)
result = list(result[0])

img_index = result.index(max(result))
emotion = label_dict[img_index]
print(emotion)
```

```python
from IPython.display import Audio
song_path = '/content/angry.mp3'
song = os.path.join(song_path,emotion)
print(f'Playing {emotion} song')
Audio('/content/angry.mp3', autoplay=True)
```
